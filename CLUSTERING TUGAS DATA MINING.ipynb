Metode Clustering

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
import seaborn as sns
```

Data Pelanggan Toko Online

```python
data = {
    'Pelanggan': ['Aidil Amal', 'Gandhi Fairuz Sidqi', 'Much Rizki Bintang Aji Pusaka', 'Rendy Ferdiansyah'],
    'Jumlah_Pesanan': [5, 3, 6, 8],
    'Total_Belanja': [120000, 100000, 50000, 110000],
    'Kunjungan': [3, 2, 5, 7]
}
```

Buat DataFrame

```python
ml = pd.DataFrame(data)
```

```python
print("Data Pelanggan:")
print(ml)
```

Fitur Untuk Clustering

```python
FF = ml[['Jumlah_Pesanan', 'Total_Belanja', 'Kunjungan']]
```

Penerapan K-Means Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=10)
ml['Cluster'] = kmeans.fit_predict(FF)
```

Tampilkan Hasil Clustering

```python
print("\n Hasil Clustering Pelanggan:")
print(ml)
```

visualisasi 2D

```python
plt.figure(figsize=(10,6))
```

```python
sns.scatterplot(data=ml, x='Jumlah_Pesanan', y='Total_Belanja', hue='Cluster', s=120, palette='Dark2')
for i in range(len(ml)):
    plt.text(ml['Jumlah_Pesanan'][i]+0.2, ml['Total_Belanja'][i], ml['Pelanggan'][i])
plt.title('Clustering Pelanggan Berdasarkan Pesanan & Belanja')
plt.xlabel('Jumlah Pesanan')
plt.ylabel('Total Belanja (Rp)')
plt.grid(True)
plt.show()
```
